## Simple Linear Regression

Simple Linear Regression models the relationship between two variables. We can predict the value of a dependent variable based on the value of an independent variable. In our data, we are interested in predicted Sleep Quality Score based on Caffeine Intake (in mg). We hope to get an equation like this:

<div align="center">
$\hat{\text{Sleep Quality Score}}_i = \hat{\beta_0} + \hat{\beta_1} (\text{Caffeine Intake}_i)$
</div>


```python
#fitting an OLS model
y = df_new['Sleep_Quality_Score']  # Dependent variable
X = sm.add_constant(df_new['Caffeine_Intake_mg'])  # Independent variable with constant added
mod = sm.OLS(y, X)  # Ordinary Least Squares model
res = mod.fit()  # Fit the model
print(res.summary())
```

After creating an OLS model we get the following model equation:

<div align="center">
$\hat{\text{Sleep Quality Score}}_i = 10.0545 - 0.0924 (\text{Caffeine Intake}_i)$
</div>

**Interpretations:**

* **$ \beta_0 = \text{10.0545}$:** The average sleep quality score is 10.0545 when Caffeine Intake is 0 (this is funny because max is 10, I guess you sleep EVEN better with no caffeine)

* **$ \beta_1 = \text{0.0924}$:** On average, Sleep Quality Score decreases by 0.0924 points with every 1 mg increase in Caffeine Intake

```python
# Scatter plot of the original data
plt.scatter(df_new['Caffeine_Intake_mg'], df_new['Sleep_Quality_Score'], color='blue', label='Data points', s=15)


predicted_values = res.predict(X)
plt.plot(df_new['Caffeine_Intake_mg'], predicted_values, color='red', label='OLS Regression Line')

# Adding labels and title
plt.title('Caffeine Intake vs Sleep Quality Score with OLS Line')
plt.xlabel('Caffeine Intake (mg)')
plt.ylabel('Sleep Quality Score')
plt.show()
```

Althought the model looks good, before we can use our linear regression model for conclusions, we must make sure we all assumptions are met. Ensuring these assumptions are met is important because violations can lead to inaccurate or misleading results.

```python
#Save residuals and fitted values to the dataframe
df_new.loc[:, 'fitted_values'] = res.fittedvalues  # Save fitted values
df_new.loc[:, 'residuals'] = res.resid  # Save residuals
df_new.head()
```

# Assumptions

## Are the residuals independant?

```python
from statsmodels.stats.stattools import durbin_watson
durbin_watson_stat = durbin_watson(res.resid)
print("Durbin-Watson Test statistic:",durbin_watson_stat)
```

For the residuals to be independent, the individual observations must not affect each other. Given that the data was simulated and not intended for the observations to depend on eachother, we can assume the residuals are independant. Using the Durbin-Watson test to verify this, we get a value close to 2 which can be interpretted as positively indicating there is dependece between residuals and therefore **the assumption is met.**

## Are any additional predictor variables required?

```python
print(f"R-squared:\t\t{res.rsquared:.3f}")
print(f"Adjusted R-squared:\t{res.rsquared_adj:.3f}")
```

The R-squared and Adjusted R-squared of 0.965 suggest that our model is explaining 96.5% of the variance in Sleep Quality Score using Caffeine Intake (mg) as the sole predictor. This is a very high value indicating that the model already performs well with just this one predictor. There are other variables included in the datset like stress levels, bedtime consistency, and light exposure hours that could also impact sleep quality, but for the analysis (and based on the R-squared values) it will be assumed that Caffeine Intake and Sleep Quality are independent of eachother, therefore **the assumption is met.**

# Transformation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox
from sklearn.preprocessing import power_transform

# Existing transformations
df_new['Caffeine_sqrt_trans'] = np.sqrt(df_new['Caffeine_Intake_mg'])
df_new['Caffeine_log_trans'] = np.log(df_new['Caffeine_Intake_mg'] + 1)  # Add 1 to avoid log(0)
df_new['Caffeine_reciprocal_trans'] = 1 / (df_new['Caffeine_Intake_mg'] + 1)  # Add 1 to avoid division by 0

df_new['Sleep_sqrt_trans'] = np.sqrt(df_new['Sleep_Quality_Score'])
df_new['Sleep_log_trans'] = np.log(df_new['Sleep_Quality_Score'] + 1)  # Add 1 to avoid log(0)
df_new['Sleep_reciprocal_trans'] = 1 / (df_new['Sleep_Quality_Score'] + 1)  # Add 1 to avoid division by 0

# Box-Cox transformation (requires all values to be positive)
df_new['Caffeine_BoxCox'], caffeine_lambda = boxcox(df_new['Caffeine_Intake_mg'] + 1)  # Adding 1 to avoid zero values
df_new['Sleep_BoxCox'], sleep_lambda = boxcox(df_new['Sleep_Quality_Score'])

# New transformations: 1/sqrt(X), X^2, sqrt(constant - X), log(constant - X)
df_new['Caffeine_inverse_sqrt_trans'] = 1 / np.sqrt(df_new['Caffeine_Intake_mg'] + 1)  # Add 1 to avoid sqrt(0)
df_new['Caffeine_squared_trans'] = df_new['Caffeine_Intake_mg'] ** 2

df_new['Sleep_inverse_sqrt_trans'] = 1 / np.sqrt(df_new['Sleep_Quality_Score'] + 1)  # Add 1 to avoid sqrt(0)
df_new['Sleep_squared_trans'] = df_new['Sleep_Quality_Score'] ** 2

# sqrt(constant - X) and log(constant - X)
constant_x = df_new['Caffeine_Intake_mg'].max() + 1  # Add a margin to avoid negative values
constant_y = df_new['Sleep_Quality_Score'].max() + 1

df_new['Caffeine_sqrt_constant_minus'] = np.sqrt(constant_x - df_new['Caffeine_Intake_mg'])
df_new['Caffeine_log_constant_minus'] = np.log(constant_x - df_new['Caffeine_Intake_mg'])

df_new['Sleep_sqrt_constant_minus'] = np.sqrt(constant_y - df_new['Sleep_Quality_Score'])
df_new['Sleep_log_constant_minus'] = np.log(constant_y - df_new['Sleep_Quality_Score'])

# Update transformation lists
x_transformations = [
    ('Caffeine_Intake_mg', 'None (X)'),
    ('Caffeine_log_trans', 'Log(X)'),
    ('Caffeine_sqrt_trans', 'Sq Root(X)'),
    ('Caffeine_reciprocal_trans', 'Reciprocal(X)'),
    ('Caffeine_inverse_sqrt_trans', 'Inv Sq Root(X)'),
    ('Caffeine_squared_trans', 'Squared(X)'),
    ('Caffeine_sqrt_constant_minus', 'Sq Root('+str(round(constant_x,1))+'-X)'),
    ('Caffeine_log_constant_minus', 'Log('+str(round(constant_x,1))+'-X)'),
    ('Caffeine_BoxCox', 'Box-Cox(X)'),
]

y_transformations = [
    ('Sleep_Quality_Score', 'None (Y)'),
    ('Sleep_log_trans', 'Log(Y)'),
    ('Sleep_sqrt_trans', 'Sq Root(Y)'),
    ('Sleep_reciprocal_trans', 'Reciprocal(Y)'),
    ('Sleep_inverse_sqrt_trans', 'Inv Sq Root(Y)'),
    ('Sleep_squared_trans', 'Squared(Y)'),
    ('Sleep_sqrt_constant_minus', 'Sq Root('+str(round(constant_y,1))+'-Y)'),
    ('Sleep_log_constant_minus', 'Log('+str(round(constant_y,1))+'-Y)'),
    ('Sleep_BoxCox', 'Box-Cox(Y)'),
]




# Function to create residual vs fitted value plots matrix
def plot_residuals_matrix(df_new):
    fig, axes = plt.subplots(len(x_transformations), len(y_transformations), figsize=(12, 12))

    for i, (x_col, x_label) in enumerate(x_transformations):
        for j, (y_col, y_label) in enumerate(y_transformations):
            X = df_new[x_col]
            y = df_new[y_col]

            try:
                # Fit OLS model
                X_const = sm.add_constant(X)
                model = sm.OLS(y, X_const).fit()

                # Plot residuals vs fitted values
                sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True,
                              ax=axes[i, j], scatter_kws={'s': 3}, line_kws={'color': 'red', 'lw': 1})

                # Remove individual titles and labels for compactness
                axes[i, j].set_xlabel('')
                axes[i, j].set_ylabel('')
                axes[i, j].tick_params(axis='both', which='major', labelsize=6)

            except Exception as e:
                print(f"Error fitting model for {x_col} and {y_col}: {e}")

    # Set row and column labels outside the plot area
    for ax, col in zip(axes[0], [y_label for _, y_label in y_transformations]):
        ax.set_title(col, size=10)

    for ax, row in zip(axes[:, 0], [x_label for _, x_label in x_transformations]):
        ax.set_ylabel(row, rotation=90, size=10, labelpad=20)

    plt.tight_layout(pad=3)
    plt.subplots_adjust(hspace=0.3, wspace=0.3)  # Adjust spacing
    plt.show()

# Call the function to create the matrix of residuals vs. fitted values plots
plot_residuals_matrix(df_new)
```

After running through all 6 assumptions necessary for a Simple Linear Regression model, all were met. But,when looking at if the residuals were normally distributed, we did find that the Shapiro-Wilk test yielded unsatisfactory results. To combat this, various transformations were tested to see if all other assumptions could still be met while the residuals became more-normally distributed. Ultimately, as seen above, no transformations had satisfactory residuals vs. fitted values plots so the untransformed data will be used for the analysis.

#Statistical Inference

Statistical inference is crucial in our analysis because it allows us to draw conclusions about a population based on a sample of data. Instead of simply describing the data at hand, statistical inference helps us determine whether the relationships we observe—such as the link between caffeine intake and sleep quality—are likely to exist in the broader population or if they could have arisen by random chance. Through methods like hypothesis testing and confidence intervals, we assess the significance, strength, and reliability of the model. For example, the p-value helps us determine whether caffeine intake is a significant predictor of sleep quality, while confidence intervals provide a range for the true effect size.

### Confidence Interval for Slope ($\beta_1$)

```python
slope_conf_int = res.conf_int().iloc[1]
print("Confidence Interval: (",slope_conf_int.iloc[0],", ",slope_conf_int.iloc[1],")", sep = "")
```

The confidence interval represents the range within which the true value of the coefficient $ \beta_1 $ is expected to lie with a certain level of confidence, 95% in for our test. The fact that the interval does not include 0 confirms that the effect of caffeine intake on sleep quality is significant and there is strong evidence that caffeine intake has a negative effect on sleep quality, rather than no effect.

  
As caffeine intake increases by 1 mg, we are 95% confident that the Sleep Quality Score will decrease by a value between -0.0938 and -0.0900 units.


### Hypothesis Test For Slope ($\beta_1$)

```python
# Extract the t-statistic and p-value for the slope (β1)
slope_t_stat = res.tvalues.iloc[1]
slope_p_value = res.pvalues.iloc[1]

# Print the test statistic and p-value
print(f"Test statistic for the slope:\t{slope_t_stat}")
print(f"P-value for the slope:\t\t{slope_p_value}")
```

The test statistic measures how many standard errors the estimated slope is away from zero. The test statistic of -96.23 is extremely large in absolute value, indicating that the slope of the relationship between caffeine intake and sleep quality score is far from zero.

The p-value represents the chance that the observed relationship between the variables Caffeine Intake and Sleep Quality could have occurred purely by random chance, assuming that there is no true relationship between them (the null hypothesis). Since the p-value is much smaller than any conventional significance level like 0.05 or 0.01, we reject the null hypothesis that there is no relationship between Caffeine Intake and Sleep Quality. This means that caffeine intake is highly statistically significant in predicting sleep quality.
