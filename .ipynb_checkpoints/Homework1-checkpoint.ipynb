## Data and Description

Energy can be produced from wind using windmills. Choosing a site for a wind farm (i.e. the location of the windmills), however, can be a multi-million dollar gamble. If wind is inadequate at the site, then the energy produced over the lifetime of the wind farm can be much less than the cost of building the operation. Hence, accurate prediction of wind speed at a candidate site can be an important component in the decision to build or not to build. Since energy produced varies as the square of the wind speed, even small errors in prediction can have serious consequences.

One possible solution to help predict wind speed at a candidate site is to use wind speed at a nearby reference site. A reference site is a nearby location where the wind speed is already being monitored and should, theoretically, be similar to the candidate site. Using information from the reference site will allow windmill companies to know the wind speed at the candidate site without going through a costly data collection period, if the reference site is a good predictor.

The Windmill data set contains measurements of wind speed (in meters per second m/s) at a **candidate site (CSpd) (column 1)** and at an accompanying **reference site (RSpd) (column 2)** for 1,116 areas. Download the Windmill.txt file from Canvas, and put it in the same folder as this Juptyer file.

#### 9. Mathematically write out the *theoretical/general* simple linear regression model for this data set (using parameters ($\beta$'s), not estimates). Do not use "x" and "y" in your model - use variable names that are fairly descriptive.

$
\text{CSpd}_i = \beta_0 + \beta_1 \cdot \text{RSpd}_i + \epsilon_i \overset{\text{iid}}{\sim} N(0, \sigma^2)
$

Where:
- $\text{CSpd}_i$ = Wind speed at the candidate site for observation $i$ (dependent variable)
- $\text{RSpd}_i$ = Wind speed at the reference site for observation $i$ (independent variable)
- $\beta_0$ = The intercept
- $\beta_1$ = The slope
- $\epsilon_i$ = The error term for observation $i$


#### 10. (a) Apply linear regression to the data, and print out a summary of the results. (b) Save the residuals and fitted values to the `wind` dataframe, and print the first few rows of the `wind` dataframe.

```python
X = df['RSpd']
y = df['CSpd']

# Add a constant to the independent variable matrix (for the intercept)
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(y, X).fit()

model.summary()
```

```python
# Add the fitted values and residuals to the DataFrame
df['Fitted'] = model.fittedvalues
df['Residuals'] = model.resid

df.head()
```

#### 11. Add the OLS regression line to the scatterplot you created in 4. Print the result.


```python
plt.figure(figsize=(6, 6))
plt.scatter(df['RSpd'], df['CSpd'], label='Data')

# Calculate the regression line
# The model's coefficients
intercept = model.params['const']
slope = model.params['RSpd']
# Generate a range of values for the reference site wind speed
RSpd_range = pd.Series([df['RSpd'].min(), df['RSpd'].max()])
# Calculate the predicted wind speeds based on the regression line
CSpd_pred = intercept + slope * RSpd_range

# Plot the regression line
plt.plot(RSpd_range, CSpd_pred, color='red', label='OLS Regression Line')

# Labels and title
plt.xlabel('Reference Site Wind Speed (RSpd) (m/s)')
plt.ylabel('Candidate Site Wind Speed (CSpd) (m/s)')
plt.title("Wind Speed Relationship: Reference vs. Candidate Site")
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

#### 12. Briefly explain the rational behind the ordinary least-squares model fit (how does OLS choose the parameter estimates?).

Ordinary Least Squares (OLS) is a way to find the best-fitting line for data points. It does this by trying to minimize the gap between the actual data points and the line it draws. These gaps are the residuals, and they are the vertical distances between the data points and the line. OLS tries to make these gaps as small as possible by adjusting the line's slope and intercept. To account for positive and negative distances, it minimizes the squared distances. The result is a line that best represents the relationship between the variables.

#### 13. Mathematically write out the *fitted* simple linear regression model for this data set using the coefficients you found above (do not use parameters/$\beta$'s). Do not use "x" and "y" in your model - use variable names that are fairly descriptive.

#### 18. Calculate the estimate of $\sigma^2$, the average squared variability of the residuals around the line. This is called the Mean Squared Error (MSE). Show your code, and print the result.

```python
# Residual Sum of Squares (RSS)
RSS = sum(model.resid**2)

# Number of observations
n = len(df)

# Estimate of sigma squared
sigma_squared = RSS / (n - 2)
sigma_squared
```
